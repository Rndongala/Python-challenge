## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)
#mouse_metadata.head()
#study_results.head()
# Combine the data into a single dataset
combine_data_df = pd.merge(mouse_metadata,study_results , on=["Mouse ID", "Mouse ID"])

# Display the data table for preview
combine_data_df
```

```python
# Checking the number of mice.
len(combine_data_df['Mouse ID'].unique())
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
#x_axix = np.arange(len(combine_data_df))
duplicate_data = combine_data_df.loc[combine_data_df.duplicated(subset=['Mouse ID', "Timepoint"]), 'Mouse ID'].unique()
duplicate_data
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_by_ID= combine_data_df.loc[combine_data_df['Mouse ID'] =='g989' ]
duplicate_by_ID
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
#duplicate_data.drop_duplicates(subset="909")
clean_dup_df= combine_data_df[combine_data_df['Mouse ID'].isin(duplicate_data)==False]
clean_dup_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.
len(clean_dup_df['Mouse ID'].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
mean = clean_dup_df.groupby('Drug Regimen').mean()['Tumor Volume (mm3)']
median = clean_dup_df.groupby('Drug Regimen').median()['Tumor Volume (mm3)']
variance = clean_dup_df.groupby('Drug Regimen').var()['Tumor Volume (mm3)']
standard_deviation = clean_dup_df.groupby('Drug Regimen').std()['Tumor Volume (mm3)']
sem = clean_dup_df.groupby('Drug Regimen').sem()['Tumor Volume (mm3)']
summary_statistics = pd.DataFrame({"Mean Tumor Volumes": mean, "Median Tumor Volumes":median, "Variance Tumor Volumes":variance, "Standard deviation Tumor Volumes":standard_deviation, "SEM Tumor Volumes":sem})

summary_statistics
# This method is the most straighforward, creating multiple series and putting them all together at the end.

```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
summary_statistics2 = clean_dup_df.groupby("Drug Regimen").agg({"Tumor Volume (mm3)":
                                                               ['mean','median','var','std','sem']})

summary_statistics2

# This method produces everything in a single groupby function
```
